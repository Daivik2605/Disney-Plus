### Signal to Noise Ratio(SNR)

```python
# In here we have used the UNCLEAN Data to find the Signal To Noise Ratio(SNR)

# Finding the total signal power by squaring the total signal
total_signal = 1450
total_signal_power = np.mean(total_signal**2)

# Find the null values of individual columns
director_noise = df['director'].isnull().sum()
cast_noise = df['cast'].isnull().sum()
country_noise = df['country'].isnull().sum()
date_added_noise = df['date_added'].isnull().sum()
rating_noise = df['rating'].isnull().sum()

# Calculate the Noise of each by squaring the noise values from the above
director_noise_power = np.mean(director_noise**2)
cast_noise_power = np.mean(cast_noise**2)
country_noise_power = np.mean(country_noise**2)
date_added_noise_power = np.mean(date_added_noise**2)
rating_noise_power = np.mean(rating_noise**2)

# Finally we get the SNR by dividing the total signal with the individual feature(Column)
director_noise_snr = total_signal_power / director_noise_power
cast_noise_snr = total_signal_power / cast_noise_power
country_noise_snr = total_signal_power / country_noise_power
date_noise_snr = total_signal_power / date_added_noise_power
rating_noise_snr = total_signal_power / rating_noise_power

# We need to make the noise data more relevant so we have converted that to the Decibels
director_snr_db = 10 * np.log10(director_noise_snr)
cast_snr_db = 10 * np.log10(cast_noise_snr)
country_snr_db = 10 * np.log10(country_noise_snr)
date_snr_db = 10 * np.log10(date_noise_snr)
rating_snr_db = 10 * np.log10(rating_noise_snr)

# Printing Statements
print("director SNR:", director_noise_snr)
print("director SNR DB:", director_snr_db)

print("\ncast SNR:", cast_noise_power)
print("cast SNR DB:", cast_snr_db)

print("\ncountry SNR:", country_noise_power)
print("country SNR DB:", country_snr_db)

print("\ndate SNR:", date_added_noise_power)
print("date SNR DB:", date_snr_db)

print("\nrating SNR:", rating_noise_power)
print("rating SNR DB:", rating_snr_db)
```
